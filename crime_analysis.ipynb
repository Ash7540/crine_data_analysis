```python
# checking top 5 'OFFENSE_CODE_GROUP' using pivot_table

offense=pd.pivot_table(crime_df.loc[:,['OFFENSE_CODE_GROUP','YEAR','OFFENSE_CODE']],index='OFFENSE_CODE_GROUP', \
              columns='YEAR',aggfunc=np.count_nonzero)
```

```python
info=pd.DataFrame(offense.apply(np.sum,axis=1))
info=info.rename(columns={0:'total'})
```

```python
sum_sort=info.sort_values(by='total',ascending=False)
top_5=sum_sort.iloc[0:5,:]
top_5
```

```python
# plotting the top 5 OFFENSE_CODE_GROUP against year

sns.set()
top_5_plot = plt.figure(figsize=(9,5))
plt.title(r'2015-2018 top 5 OFFENSE_CODE_GROUP horizontal bar plot')
plt.barh(range(top_5.index.shape[0]), top_5.loc[:, 'total'])
plt.ylabel('OFFENSE_CODE_GROUP')
plt.yticks(range(top_5.index.shape[0]), top_5.index)
y = np.arange(top_5.index.shape[0])
x = np.array(top_5['total'])
for i,j in zip(x,y):
    plt.text(i,j,'%d'%i,ha='center')
top_5_plot.savefig('./top5_group_bar.png')
plt.show()
```

# ***PCA: - Principle Component Analysis***

```python
from sklearn.preprocessing import LabelEncoder
df=crime_df.iloc[:,:]
df.loc[:,'OFFENSE_CODE_GROUP']=LabelEncoder().fit_transform(df.loc[:,'OFFENSE_CODE_GROUP'])
df.loc[:,'OFFENSE_DESCRIPTION']=LabelEncoder().fit_transform(df.loc[:,'OFFENSE_DESCRIPTION'])
df.loc[:,'DISTRICT']=LabelEncoder().fit_transform(df.loc[:,'DISTRICT'].astype('str'))
df.loc[:,'SHOOTING']=LabelEncoder().fit_transform(df.loc[:,'SHOOTING'].astype('str'))
df.loc[:,'DAY_OF_WEEK']=LabelEncoder().fit_transform(df.loc[:,'DAY_OF_WEEK'])
df.loc[:,'UCR_PART']=LabelEncoder().fit_transform(df.loc[:,'UCR_PART'].astype('str'))
df.loc[:,'STREET']=LabelEncoder().fit_transform(df.loc[:,'STREET'].astype('str'))
```

```python
df.head()
```

```python
df.loc[:,'REPORTING_AREA']=LabelEncoder().fit_transform(df.loc[:,'REPORTING_AREA'])
```

```python
df.head()
```

```python
data=df.loc[:,['OFFENSE_CODE', 'OFFENSE_CODE_GROUP','OFFENSE_DESCRIPTION', \
                 'DISTRICT', 'REPORTING_AREA', 'SHOOTING', 'YEAR','DAY_OF_WEEK', 'HOUR', 'UCR_PART', 'STREET']]
target=df.loc[:,'MONTH']
```

```python
from sklearn.decomposition import PCA
pcamodel=PCA(n_components=11).fit(data)
```

```python
print(pcamodel.explained_variance_ratio_)
```
